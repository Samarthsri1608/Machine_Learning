```python
tips = sns.load_dataset("tips")
```

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex")
```

```python
tips
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", dodge=True)
```

```python
sns.pairplot(data=tips)
```

```python
sns.barplot(x="day", y="tip", data=tips, hue="sex")
```

```python
tips['sex_bool'] = 0
tips.loc[tips['sex'] == 'Male', 'sex_bool'] = 1
```

```python
tips[['sex','sex_bool']]
```

```python
tips['smoker_bool'] = 0
tips.loc[tips['smoker'] == 'Yes', 'smoker_bool'] = 1
```

```python
tips
```

```python
tips.drop(["sex", "smoker"], axis=1)
```

```python
day_map = {'Mon': 1, 'Tue': 2, 'Wed': 3, 'Thur': 4, 'Fri': 5, 'Sat': 6, 'Sun': 7}

tips['day_num'] = tips['day'].map(day_map)
tips
```

```python
time_map = {'Breakfast': 1, 'Lunch': 2, 'Dinner':3}

tips['time_num'] = tips['time'].map(time_map)
tips
```

```python
tips.drop(["sex", "smoker", "day", "time"], axis=1)
```

```python
corr_mat = tips[["total_bill","tip","size","sex_bool","smoker_bool","day_num","time_num"]].corr()
sns.heatmap(corr_mat, annot=True, cmap="YlGnBu")
```
